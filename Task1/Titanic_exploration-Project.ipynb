```python
new_data = data.copy() # Make another copy of the new dataframe

#Changing the embarked keys to readable ones
new_data['Embarked'] = new_data['Embarked'].replace( ['C', 'S', 'Q'],  ['Cherbourg', 'Southampton', 'Queenstown'] )

#Changing the survived keys to readable ones
new_data['Survived'] = new_data['Survived'].replace( [0, 1],   ['No', 'Yes'] )

new_data
```

```python
cross_tab_count = pd.crosstab(new_data['Pclass'], new_data['Survived'])

ct_percent = pd.crosstab(new_data['Pclass'], new_data['Survived'], normalize='index') * 100

# Combining both (counts + percentages)
ct_combined =  cross_tab_count.astype(str) + " (" + ct_percent.round(1).astype(str) + "%)"

print("Counts with Percentages by Passenger Class:")
print(ct_combined)
```

Summary: The cross-tabulation reveals a clear class-based disparity in survival outcomes. First-class passengers had the highest survival rate (63%), second-class passengers had moderate chances (47%), while third-class passengers had the lowest survival rate (24%). This indicates that socio-economic status, represented by passenger class, played a significant role in survival during the Titanic disaster.

```python
crosstab_data = {
    0: [80, 97, 372],   # Did Not Survive
    1: [136, 87, 119]   # Survived
}

crosstab = pd.DataFrame(crosstab_data, index=[1, 2, 3])
crosstab.index.name = "Pclass"
crosstab.columns = ["Did Not Survive", "Survived"]

# Convert counts to percentages by row
crosstab_percent = crosstab.div(crosstab.sum(axis=1), axis=0) * 100

# Stacked bar chart
crosstab_percent.plot( kind="bar", stacked=True, figsize=(8, 6),
     color=["black", "blue"]  # orange for no, green for yes
)

plt.title("Survival Rates by Passenger Class")
plt.ylabel("Percentage")
plt.xlabel("Passenger Class")
plt.legend(title="Outcome")
plt.xticks(rotation=0)
plt.show()
```

```python
# Impute missing numerical values
new_data['Age'].fillna(df['Age'].median())   # median for Age

# Impute missing categorical value
new_data['Embarked'].fillna(df['Embarked'].mode()[0])  # mode for Embarked

# Verify missing values are handled
print(new_data.isnull().sum())
```

```python
# Filter children (Age < 18)
children = new_data[new_data['Age'] < 18]

# Count number of children who survived vs did not survive
survival_counts = children['Survived'].value_counts().sort_index()  # 0 = No, 1 = Yes

# Bar chart
plt.bar(['No', 'Yes'], survival_counts, color=['black', 'blue'])
plt.title('Survival Count for Children (Age < 18)')
plt.xlabel('Survival Status')
plt.ylabel('Number of Children')
plt.show()

# Optional: print counts
print(survival_counts)
```

Summary: Children below 18 years of age have higher chances of surviving.

```python
survival_by_class = children.groupby('Pclass')['Survived'].value_counts().unstack(fill_value=0)

survival_by_class.plot(kind='bar', stacked=False, color=['black', 'blue'], figsize=(8,6))
plt.title('Survival of Children by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Children')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.xticks(rotation=0)
plt.show()
```

```python
# Creating FamilySize column
new_data['FamilySize'] = new_data['SibSp'] + new_data['Parch'] + 1  # +1 to include the passenger themselves

# categorizing family sizes
def family_category(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'Small Family'
    else:
        return 'Large Family'

new_data['FamilyCategory'] = new_data['FamilySize'].apply(family_category)

# Survival count by family category
sns.countplot(x='FamilyCategory', hue='Survived', data=new_data, palette=['black','blue'])
plt.title('Survival by Family Size')
plt.xlabel('Family Category')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No','Yes'])
plt.show()
```

Summary: The Survival rate of passengers travelling 'Alone' is higher than passengers in Large/ Small family.

```python
# Creating FamilySize column
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1  # +1 to include the passenger themselves

# categorizing family sizes
def family_category(size):
    if size == 1:
        return 'Alone'
    elif size <= 4:
        return 'Small Family'
    else:
        return 'Large Family'

df['FamilyCategory'] = df['FamilySize'].apply(family_category)

# Extract 'Title' from 'Name' using a raw string
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Replace rare titles with a common one
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df['Title'] = df['Title'].replace({'Mlle':'Miss', 'Ms':'Miss', 'Mme':'Mrs'})

# Create 'FarePerPerson'
df['FarePerPerson'] = df['Fare'] / df['FamilySize']

print("New features created:")
display(df[['FamilySize', 'FamilyCategory', 'Title', 'FarePerPerson']].head())
```

```python
# Extract title from Name 
df['Title'] = df['Name'].str.extract(r' ([A-Za-z]+)\.', expand=False)

# Replace rare titles with 'Rare'
df['Title'] = df['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

# Standardize common variations
df['Title'] = df['Title'].replace({'Mlle':'Miss', 'Ms':'Miss', 'Mme':'Mrs'})

df[['Name','Title']].head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='Title', hue='Survived', data=df, palette=['black','blue'])
plt.title('Survival by Title')
plt.xlabel('Title')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No','Yes'])
plt.xticks(rotation=45)
plt.show()
```

```python
port_mapping = {'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}
df['EmbarkedFull'] = df['Embarked'].map(port_mapping)

# Count plot: survival by port of embarkation
sns.countplot(x='EmbarkedFull', hue='Survived', data=df, palette=['black','blue'])
plt.title('Survival Count by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No','Yes'])
plt.show()

# percentage survival per port
embarked_survival = df.groupby('Embarked')['Survived'].mean() * 100
print("Survival percentage by Embarked:")
print(embarked_survival)
```

```python
# Survival by Class + Port
# Map codes to full port names
port_mapping = {'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'}
df['EmbarkedFull'] = df['Embarked'].map(port_mapping)

# Group by Pclass and full port names, calculate survival rate
class_embarked_survival = df.groupby(['Pclass', 'EmbarkedFull'])['Survived'].mean().unstack()

# Plot as a heatmap
plt.figure(figsize=(8,5))
sns.heatmap(class_embarked_survival, annot=True, fmt=".2f", cmap="Blues")
plt.title('Survival Rate by Class and Port of Embarkation')
plt.ylabel('Passenger Class')
plt.xlabel('Port of Embarkation')
plt.show()
```

```python
# Deck analysis:
# Extracting deck letter (first character of Cabin)
df['Deck'] = df['Cabin'].str[0]

# Fill missing values with 'Unknown'
df['Deck'] = df['Deck'].fillna('Unknown')

# Count plot: survival by deck
plt.figure(figsize=(10,5))
sns.countplot(x='Deck', hue='Survived', data=df, palette=['black','blue'], order=sorted(df['Deck'].unique()))
plt.title('Survival by Deck')
plt.xlabel('Deck')
plt.ylabel('Number of Passengers')
plt.legend(title='Survived', labels=['No','Yes'])
plt.show()

# survival rate per deck
deck_survival = df.groupby('Deck')['Survived'].mean() * 100
print("Survival percentage by Deck:")
print(deck_survival.sort_values(ascending=False))
```
